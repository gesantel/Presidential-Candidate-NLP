```python
#Add word count
df = data.copy()
df["word_count"] = df["text"].str.split().str.len()
```

```python
#Small sentiment langauage processing model 
model_name = 'azizbarank/distilroberta-base-sst2-distilled'
access_token = REMOVED
tok = AutoTokenizer.from_pretrained(model_name, use_auth_token=Removed)
model = AutoModelForSequenceClassification.from_pretrained(Removed)



sentiment = pipeline(
    "sentiment-analysis",
    model=model,
    tokenizer=tok,
    device=-1 
)
def add_sentiment_columns_chunked(df, text_col="text", batch_size=16, chunk_size=64):
    labels = []
    scores = []
    # Ensure text is string
    texts_all = df[text_col].fillna("").astype(str).tolist()
    # Process in chunks
    for i in range(0, len(texts_all), chunk_size):
        chunk_texts = texts_all[i:i+chunk_size]
        results = sentiment(chunk_texts, batch_size=batch_size, truncation=True)
        labels.extend([r['label'] for r in results])
        scores.extend([r['score'] for r in results])
    df['label'] = labels
    df['score'] = scores
    return df

df = add_sentiment_columns_chunked(df)
def get_by_speaker(speaker):
    return df[df['speaker'] == speaker]

df.speaker.unique()
```

```python
#Use python collections.counter counter.most_common to find most common words by speaker
def top_words(texts, n=10):
    texts = texts.dropna().astype(str) #drop NA and ensure type string
    words = re.findall(r'\b\w+\b', ' '.join(texts).lower()) #identify words with regex, join into one string, lowercase
    common = Counter(words).most_common(n) #returns {word:count}, grabs 10 most common words
    return pd.DataFrame(common, columns=['word','count']) #returns data as dataframe

ddf = df.groupby('speaker')['text'].apply(lambda x: top_words(x, 10)).reset_index(level =1, drop = True).reset_index()
#pd.set_option('display.max_rows', None)

ddf
```

```python
#Linguistic feature analysis by speaker. Self vs collective statements
df['text_lower'] = df['text'].astype(str).str.lower().str.strip()

for word in ['i','we','he']:
    df[word + '_count'] = df['text_lower'].str.count(rf'\b{word}\b') #regex to identify we, i, he statements, save as columns

counts = df.groupby('speaker')[['we_count', 'i_count', 'he_count']].sum().sort_index()
print(counts)
```

```python
#We normalize per 1000 words

for word in ['he', 'we', 'i']:
    df[word+'_per_1000'] = 1000*df[word+'_count']/(df['word_count']+1)

pronoun_means = df.groupby('speaker')[['he_per_1000', 'we_per_1000','i_per_1000']].mean()
pronoun_means
```

```python
pronoun_means.plot(kind = 'bar', figsize = (10,6))
plt.title('Pronoun usage per 1000 words by Speaker')
plt.ylabel('Occurences per 1,000 words.')
plt.xlabel('Speaker')
plt.xticks(rotation = 45)
plt.legend(title = 'Pronoun')
plt.tight_layout()
plt.show
```
